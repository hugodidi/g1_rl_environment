```python
# ============================================================
# Adapting the MuJoCo Simulator for ROS 2 Integration
# ============================================================
#
# In this project, the default simulator could not run because:
#   - Docker was not available,
#   - some dependencies were incompatible,
#   - and the original environment expected a containerized setup.
#
# To solve this, I manually integrated MuJoCo inside my Python
# virtual environment and implemented a ROS 2 bridge.
#
# ------------------------------------------------------------
# 1) Using MuJoCo inside the virtual environment
# ------------------------------------------------------------
# MuJoCo is installed and executed directly inside the virtual
# environment (instead of Docker). This allows:
#   - running the G1 simulation from Python,
#   - exchanging data with ROS 2 topics,
#   - controlling the robot through /mujoco/joint_torque_cmd.
#
# ------------------------------------------------------------
# 2) Creating the deploy_mujoco_ros module
# ------------------------------------------------------------
# I created the directory:
#   unitree_rl_gym/deploy/deploy_mujoco/
#
# Inside it, I added:
#   deploy_mujoco_ros.py
#
# This file works as a real ROS 2 simulation node. It:
#   - loads the MuJoCo model,
#   - runs the physics engine,
#   - applies PD control,
#   - publishes JointState messages,
#   - receives RL actions from ROS.
#
# ------------------------------------------------------------
# 3) Adapting the scene file to a 29-DOF G1 robot
# ------------------------------------------------------------
# The default scene.xml does not match the RL policy (29 actions).
# I created a new file:
#   scene_29.xml
#
# located inside:
#   unitree_rl_gym/resources/robots/g1_description/
#
# This scene includes:
#   g1_29dof.xml
#
# ensuring perfect matching between:
#   - the policy (29 actions),
#   - the MuJoCo model (29 actuated joints).
#
# ------------------------------------------------------------
# 4) Exporting the RSL-RL checkpoint to ONNX
# ------------------------------------------------------------
# To run the trained policy through ONNX (instead of PyTorch),
# I created the script:
#   ~/RL/export_rslrl_to_onnx_from_ckpt.py
#
# This script:
#   - loads an RSL-RL checkpoint,
#   - exports the model to ONNX,
#   - allows running inference inside the ROS 2 node using
#     onnxruntime.
#
# With this, the control loop becomes:
#
#   MuJoCo → publishes /mujoco/joint_states
#   ONNX policy → subscribes, computes actions
#   MuJoCo → receives /mujoco/joint_torque_cmd
#
# ------------------------------------------------------------
# Final Result
# ------------------------------------------------------------
# The entire RL → Simulation → ROS 2 pipeline works without Docker.
# We now have:
#   ✓ a MuJoCo simulator integrated in Python,
#   ✓ a ROS 2 bridge node,
#   ✓ a consistent 29-DOF G1 model,
#   ✓ ONNX-based real-time policy execution.
#
# This setup is useful for teaching:
#   - reinforcement learning deployment,
#   - robot simulation,
#   - ROS 2 integration and real-time control.
#
# ============================================================

cd ~/ros2rob_ws
colcon build --packages-select motion_tracking_controller
```

```python
# ================================
#   Summary of the Issue & Fix
# ================================

# 1. The Problem
# --------------
# The BeyondMimic project (motion_tracking_controller) was designed for:
#   - Ubuntu 24.04
#   - ROS 2 Jazzy
#   - the Unitree "legged_control2" stack
#
# On Ubuntu 22.04 with ROS 2 Humble:
#   - the robot immediately falls in MuJoCo
#   - controllers never load
#   - repeated messages:
#       "Could not contact /controller_manager/list_controllers"
#   - plugin loading errors:
#       "Could not find library mujoco_ros2_control::MujocoRos2ControlPlugin"
#
# This is expected: ROS 2 Humble originally did NOT include
# the G1 control stack nor the proper MuJoCo integration needed
# for BeyondMimic.

# 2. Working Solution (for ROS 2 Humble)
# --------------------------------------
# Instead of upgrading to Ubuntu 24 + Jazzy or using Docker,
# I use precompiled ROS 2 Humble packages:
#
#   ros-humble-mujoco-ros2-control
#   ros-humble-mujoco-sim-ros2
#   ros-humble-unitree-description
#   ros-humble-unitree-sdk1-ament
#   ros-humble-unitree-sdk2-ament
#   ros-humble-unitree-systems
#
# These packages provide:
#   - the correct mujoco_ros2_control plugin for Humble
#   - the Unitree G1 robot description
#   - the Unitree ros2_control interfaces
#
# Installation is done in two steps:
#
#   1) ./install_g1_stack.sh
#      -> installs the full Unitree G1 stack + MuJoCo integration
#         through APT for ROS 2 Humble
#
#   2) ./setup_g1_workspace.sh
#      -> creates ~/colcon_ws, clones unitree_bringup and
#         motion_tracking_controller, installs dependencies,
#         and builds only the necessary packages.
#
# IMPORTANT:
#   source:
#       source /opt/ros/humble/setup.zsh
#       source ~/colcon_ws/install/setup.zsh

# 3. Running the BeyondMimic Policy in MuJoCo
# -------------------------------------------
# To run the trained policy in simulation:
#
#   ros2 launch motion_tracking_controller mujoco.launch.py \
#       wandb_path:=<my_project/my_policy_id>
#
# This launch file:
#   - downloads the ONNX policy from Weights & Biases
#   - generates a temporary controllers.yaml
#   - starts MuJoCo + ros2_control
#   - launches state_estimator + walking_controller
#
# Result: the G1 humanoid executes the BeyondMimic motion
# correctly inside MuJoCo on ROS 2 Humble.

# ================================
# End of Summary 
# ================================

# G1 Stack Installation Process
sudo ./install_g1_stack.sh

# WS setup
sudo ./setup_g1_workspace.sh

# Or in one script
 sudo ./install_unitree_beyondmimic.sh


 # Sourcing
source /opt/ros/humble/setup.zsh
source ~/colcon_ws/install/setup.zsh


python3 -m pip install --user --upgrade pip
python3 -m pip install --user wandb

#Good trained policy
ros2 launch motion_tracking_controller mujoco.launch.py wandb_path:=jmayb-universidad-de-le-n/horse_stance_punch/8aqcnhm7

#Long performance
ros2 launch motion_tracking_controller mujoco.launch.py wandb_path:=jmayb-universidad-de-le-n/p_dance1_subject3/fs4onwac

# Bad trained policy
ros2 launch motion_tracking_controller mujoco.launch.py wandb_path:=jmayb-universidad-de-le-n/hooks_punch/am8zy8tb




source ~/.zshrc   
env | grep -E 'CYCLONEDDS|RMW'
cd ~/colcon_ws
colcon build --symlink-install --cmake-args -DCMAKE_BUILD_TYPE=RelWithDebInfo --packages-up-to unitree_bringup
colcon build --symlink-install --cmake-args -DCMAKE_BUILD_TYPE=RelWithDebInfo --packages-up-to motion_tracking_controller

ros2 launch motion_tracking_controller real.launch.py network_interface:=eno1 wandb_path:=jmayb-universidad-de-le-n/horse_stance_punch/8aqcnhm7




# Run inside the Docker
# Install it and others components
# Build and make it runing

cd ~/RL/g1_docker

sudo docker compose build
sudo docker compose up -d
sudo docker exec -it ros_humble_g1_beyondmimic bash

source /opt/ros/humble/setup.bash
source /colcon_ws/install/setup.bash

ros2 launch motion_tracking_controller mujoco.launch.py \
  wandb_path:=jmayb-universidad-de-le-n/horse_stance_punch/8aqcnhm7

# In case of error, runing the simulation due to :
# Authorization required, but no authorization protocol specified
# ERROR: could not initialize GLFW
xhost +local:

# 1. Authorize the local user ( Docker)
xhost +local:

# 2. Re-start the container
cd ~/RL/g1_docker
docker compose down
docker compose up -d

# 3. Login and run again
docker exec -it ros_humble_g1_beyondmimic bash
source /opt/ros/humble/setup.bash
source /colcon_ws/install/setup.bash

ros2 launch motion_tracking_controller mujoco.launch.py \
  wandb_path:=jmayb-universidad-de-le-n/horse_stance_punch/8aqcnhm7
```
